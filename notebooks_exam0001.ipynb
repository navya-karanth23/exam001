```python
fruits=['banana','apple','orange','mango']
def has_g(fruits):
    return 'g' in fruits
filtered_fruits=list(filter(has_g,fruits))
print(filtered_fruits)
```

```python
from functools import reduce
nums1=[1,2,3,4,5]
summ=reduce(lambda x,y:x+y,nums1)
print(summ)
```
